### Qu es distplot?

- Trazar de forma flexible una distribucin univariada de las observaciones.
- Esta funcin combina la funcin hist de matplotlib (con el clculo automtico de un buen tamao de contenedor predeterminado) con las funciones kdeplot () y rugplot (). Tambin puede ajustarse a las distribuciones de scipy.stats y trazar el PDF estimado sobre los datos.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
num = np.random.randn(150)
sns.distplot(num,color ='green')
```

```python
label_dist = pd.Series(num,name = " Variable x")
sns.distplot(label_dist,color = "red")
```

```python
# Plot the distribution with a kenel density. estimate and rug plot:

sns.distplot(label_dist,hist = False,color = "red")
```

```python
# Plot the distribution with a kernel density estimate and rug plot:

sns.distplot(label_dist,rug = True,hist = False,color = "red")
```

```python
# Plot the distribution with a histogram and maximum likelihood gaussian distribution fit:

from scipy.stats import norm
sns.distplot(label_dist, fit=norm, kde=False)
```

```python
sns.distplot(label_dist, vertical =True)
```

### Qu es boxplot?

- Dibujar un diagrama de caja para mostrar las distribuciones con respecto a las categoras.
- Un diagrama de caja (o diagrama de caja y bigotes) muestra la distribucin de datos cuantitativos de una manera que facilita las comparaciones entre variables o a travs de los niveles de una variable categrica.
- El cuadro muestra los cuartiles del conjunto de datos, mientras que los bigotes se extienden para mostrar el resto de la distribucin, excepto por los puntos que se determinan como "valores atpicos" utilizando un mtodo que es una funcin del rango intercuartil.

```python
#tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style("whitegrid")
sns.boxplot(x = tips['size'])
```

```python
sns.set_style("whitegrid")
sns.boxplot(x = tips['total_bill'])
```

```python
# Draw a vertical boxplot grouped by a categorical variable:

sns.set_style("whitegrid")
sns.boxplot(x = 'day',y = 'total_bill',data = tips)
```

```python
# Draw a boxplot with nested grouping by two categorical variables:

sns.set_style("whitegrid")
sns.boxplot(x = 'day',y = 'total_bill',hue = 'smoker',data = tips,palette = 'Set3')
```

```python
sns.set_style("whitegrid")
sns.boxplot(x = 'day',y = 'total_bill',hue = 'time',data = tips)
```

```python
# Draw a boxplot with nested grouping when some bins are empty:

sns.set_style("whitegrid")
sns.boxplot(x = 'day',y = 'total_bill',hue = 'time',data = tips,linewidth = 3.5)
```

```python
# Draw a boxplot for each numeric variable in a DataFrame:
iris = sns.load_dataset("iris")
sns.boxplot(data = iris,orient ='h',palette = 'Set2')

```

```python
# Use swarmplot() to show the datapoints on top of the boxes:
sns.boxplot(x = 'day',y = 'total_bill',data = tips)
sns.swarmplot(x = 'day',y = 'total_bill',data = tips,color = 'black')
```

### Qu es stripplot?
- Dibujar un diagrama de dispersin donde una variable es categrica.
- Se puede dibujar una grfica de franjas por s misma, pero tambin es un buen complemento para una trama de violn o caja en los casos en que desee mostrar todas las observaciones junto con alguna representacin de la distribucin subyacente.
- http://seaborn.pydata.org/generated/seaborn.stripplot.html?highlight=stripplot#seaborn.stripplot

```python
# load tips dataset

tips_data = pd.read_csv("../data/tips.csv")
tips_data.head()
```

```python
# Draw a single horizontal strip plot:

sns.set_style("whitegrid")
#tips_data = sns.load_dataset("tips")
sns.stripplot(x = tips_data["total_bill"],size = 4,color = "green")
```

```python
sns.stripplot(x = 'day',y ='total_bill',data = tips_data,size =7,order = ['Fri','Sat','Sun','Thur'])
```

```python
# use of jitter

sns.stripplot(x = 'day',y ='total_bill',data = tips_data,size =7,order = ['Fri','Sat','Sun','Thur'],jitter = True)
```

```python
# Draw horizontal strips:
sns.stripplot(x="total_bill", y="day", data=tips_data,jitter = True)
```

```python
# Draw outlines around the points:
sns.stripplot(x="total_bill", y="day", data=tips_data,jitter = True,linewidth = 1)
```

```python
# Nest the strips within a second categorical variable:

sns.stripplot(x="day", y="total_bill", data=tips_data,jitter = True,hue = "sex")
plt.legend(loc=3)
```

```python
# Draw each level of the hue variable at different locations on the major categorical axis:
sns.stripplot(x="day", y="total_bill", data=tips_data,jitter = True,hue = "sex",dodge = True)
plt.legend()
```

```python
# Control strip order by passing an explicit order:

sns.stripplot(x="time", y="tip", data=tips_data,order = ["Dinner","Lunch"],size = 8)

```

### Dibuja tiras de observaciones en la parte superior de un diagrama de caja:

```python
sns.boxplot(x="tip", y="day", data=tips_data, whis=np.inf)
sns.stripplot(x="tip", y="day", data=tips_data,jitter=True, color=".3")
```

### Dibuja tiras de observaciones en la parte superior de un plot de violn:

```python
sns.boxplot(x="day", y="total_bill", data=tips_data, whis=np.inf)
sns.stripplot(x="day", y="total_bill", data=tips_data, jitter=True)
sns.violinplot(x="day", y="total_bill", data=tips_data,inner=None, color=".8")
```
